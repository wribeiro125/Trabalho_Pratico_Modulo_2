```
#Importa as bibliotecas utilizadas nesta análise
import numpy as np
import pandas as pd
```

```
#Carrega o dataset
bike_df = pd.read_csv('https://pycourse.s3.amazonaws.com/bike-sharing.csv')
```

```
#Exibe as 5 primeiras linhas do Data Frame
bike_df.head()
```

```
#Exibe algumas informações relevantes para a análise
bike_df.info()
```

```
#1 - Qual o tamanho desse dataset (número de linhas, número de colunas)?
bike_df.shape
```

```
#Exibe um resumo estatístico do Data Frame
bike_df.describe()
```

```
#2 - Qual a média da coluna windspeed?
bike_df['windspeed'].mean()
```

```
#3 - Qual a média da coluna temp?
bike_df['temp'].mean()
```

```
#Transformando a coluna no tipo date
bike_df['datetime'] = pd.to_datetime(bike_df['datetime'])
```

```
#4 - Quantos registros de locações existem para o ano de 2011 (número de linhas, número de colunas)?
print((bike_df.datetime.dt.year == 2011).sum())
```

```
#5 - Quantos registros de locações existem para o ano de 2012 (número de linhas, número de colunas)?
print((bike_df.datetime.dt.year == 2012).sum())
```

```
#6 - Quantas locações de bicicletas foram efetuadas em 2011?
bike_df_2011 = bike_df[bike_df.datetime.dt.year == 2011]
bike_df_2011['total_count'].sum()
```

```
#7 - Quantas locações de bicicletas foram efetuadas em 2012?
bike_df_2012 = bike_df[bike_df.datetime.dt.year == 2012]
bike_df_2012['total_count'].sum()
```

```
#8 - Qual estação do ano contém a maior média de locações de bicicletas?
#9 - Qual estação do ano contém a menor média de locações de bicicletas?
bike_df_season = bike_df.groupby('season').total_count.mean()
bike_df_season
```

```
#10 - Qual horário do dia contém a maior média de locações de bicicletas?
bike_df_hour = bike_df.groupby('hour')['total_count'].mean().sort_values(ascending=False)
bike_df_hour[:1] #primeiro valor
```

```
#11 - Qual horário do dia contém a menor média de locações de bicicletas?
bike_df_hour = bike_df.groupby('hour')['total_count'].mean().sort_values(ascending=False)
bike_df_hour[-1:] #último valor
```

```

bike_df_hour = bike_df.groupby('hour')['total_count'].mean()
bike_df_hour.min()
```

```
#12 - Que dia da semana contém a maior média de locações de bicicletas?
bike_df_weekday = bike_df.groupby('weekday')['total_count'].mean().sort_values(ascending=False)
bike_df_weekday[:1] #Primeiro valor
```

```
#13 - Que dia da semana contém a menor média de locações de bicicletas?
bike_df_weekday = bike_df.groupby('weekday')['total_count'].mean().sort_values(ascending=False)
bike_df_weekday[-1:] #Último valor
```

```
#14 - Às quartas-feiras (weekday = 3), qual o horário do dia contém a maior média de locações de bicicletas?
bike_df_weekday_3 = bike_df[bike_df.weekday == 3].groupby('hour')['total_count'].mean().sort_values(ascending=False)
bike_df_weekday_3[:1] #Primeiro valor
```

```
#15 - Aos sábados (weekday = 6), qual o horário do dia contém a maior média de locações de bicicletas?
bike_df_weekday_6 = bike_df[bike_df.weekday == 6].groupby('hour')['total_count'].mean().sort_values(ascending=False)
bike_df_weekday_6[:1] #Primeiro valor
```
